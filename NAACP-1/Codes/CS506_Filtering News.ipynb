```python
globe2014_1 = pd.read_csv("Sentiment_Globe2014_1.csv")
globe2014_2 = pd.read_csv("Sentiment_Globe2014_2.csv")
globe2014 = globe2014_1.append(globe2014_2, ignore_index = True)
globe2014.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
```

```python
globe2017 = pd.read_csv("Sentiment_Globe2017_1.csv")
globe2017.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
globe2018_1 = pd.read_csv("Sentiment_Globe2018_1.csv")
globe2018_2 = pd.read_csv("Sentiment_Globe2018_2.csv")
globe2018_3 = pd.read_csv("Sentiment_Globe2018_3.csv")
globe2018_mid = globe2018_1.append(globe2018_2, ignore_index = True)
globe2018 = globe2018_mid.append(globe2018_3, ignore_index = True)
globe2018.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
def getNbh(s):

    neigh = ['Allston', 'Back Bay', 'Bay Village', 'Beacon Hill', 'Brighton', 'Charlestown', 
             'Dorchester', 'Downtown', 'East Boston', 'Fenway', 'Harbor Islands', 'Hyde Park',
             'Jamaica Plain', 'Longwood Medical Area', 'Mattapan', 'Mission Hill', 'North End', 
             'Roslindale', 'Roxbury', 'South Boston', 'South Boston Waterfront', 'South End', 
             'West End', 'West Roxbury', 'Chelsea', 'Revere', 'Winthrop'] 

    nbh = []

    for neighborhood in neigh:
        if bool(re.search(neighborhood, s)):
            nbh.append(neighborhood)
        if bool(re.search(neighborhood.upper(), s)):
            nbh.append(neighborhood)

    return nbh
```

```python
def getSubNbh(s):

    sub_neigh = ['Chelsea', 'East Boston', 'Franklin Park', 'Orient Heights', 'Revere', 'Ashmont', 
                 'Back Bay', 'Bay Village', 'Beacon Hill', 'Boston', 'Cedar Grove', 'City Point', 
                 'Dock Square', 'Dorchester', 'Dorchester Center', 'Dorchester Heights', 'Downtown Crossing',
                 'Fields Corner', 'Forest Hills', 'Fort Independence', 'Fort Winthrop', 'Grove Hall',
                 'Jamaica Plain', 'Jeffries Point', 'Lower Mills', 'Mattapan', 'Meeting House Hill',
                 'Mount Bowdoin', 'Mount Ida', 'Neponset', 'North End', 'Roxbury', 'Roxbury Crossing',
                 'Savin Hill', 'South Boston', 'Uphams Corner', 'Cottage Hill', 'Cottage Park',
                 'Court Park', 'Fort Dawes', 'Fort Warren', 'Point Shirley', 'Beachmont', 'Crescent Beach',
                 'Oak Island', 'Point of Pines', 'Suffolk Downs Station', 'Winthrop', 'Winthrop Beach',
                 'Winthrop Highlands', 'Aberdeen', 'Allston', 'Bellevue', 'Brighton', 'Clarendon Hills',
                 'Faneuil', 'Germantown', 'Highland', 'Hyde Park', 'Mount Hope', 'North Brighton',
                 'Roslindale', 'Stonybrook Village', 'West Roxbury', 'Readville', 'Egypt Beach', 
                 'Charlestown', 'Chinatown', 'Fairmount', 'Mishawum', 'Totant', 'Winnisimmet'] 

    sub_nbh = []

    for neighborhood in sub_neigh:
        if bool(re.search(neighborhood, s)):
            sub_nbh.append(neighborhood)
        if bool(re.search(neighborhood.upper(), s)):
            sub_nbh.append(neighborhood)

    return sub_nbh
```

```python
# bostonglobe2014
nbh_all = []
sub_nbh_all = []

for i in range(len(globe2014)):
    story = globe2014['Sentence'][i]
    globe2014['Sentence'][i] = story.replace('\\n\\t','').replace('\\t','').replace('\\n','')   
    nbh = getNbh(globe2014['Sentence'][i])
    sub_nbh = getSubNbh(globe2014['Sentence'][i])
    
    nbh_all.append(nbh)
    sub_nbh_all.append(sub_nbh)
    
globe2014['Neighborhood'] = nbh_all
globe2014['Sub-neighborhood'] = sub_nbh_all
```

```python
# drop unusable data
print(len(globe2014))

drop_list = []
for i in range(len(globe2014)):
    if len(globe2014['Sub-neighborhood'][i]) > 4 or len(globe2014['Sub-neighborhood'][i]) == 0 or globe2014['Sub-neighborhood'][i] == ['Boston']:
        drop_list.append(i)
        
print(len(drop_list))
globe2014_cleaned = globe2014.drop(drop_list)
globe2014_cleaned.reset_index(inplace=True, drop=True)
print(len(globe2014_cleaned))
```

```python
black_nbh = ['Roxbury', 'Grove Hall', 'Mount Ida', 'Dorchester', 'Mount Bowdoin', 'Ashmont', 'Lower Mills', 
             'Mattapan', 'Germantown', 'Hyde Park', 'Forest Hills']

check_black = []
for i in range(len(globe2014_cleaned)):
    tags = globe2014_cleaned['Sub-neighborhood'][i]
    count = 0
    for tag in tags:
        if tag in black_nbh:
            count += 1
        
    if count >= 2:
        check_black.append('Y')
    else:
        check_black.append('N')

        
globe2014_cleaned['Black or Not [Y/N]'] = check_black
globe2014_cleaned      
```

```python
globe2014_cleaned.drop_duplicates(subset=['Sentence'], inplace=True)
globe2014_cleaned
```

```python
globe2014_black = globe2014_cleaned.loc[globe2014_cleaned['Black or Not [Y/N]'] == 'Y']
globe2014_black
```
