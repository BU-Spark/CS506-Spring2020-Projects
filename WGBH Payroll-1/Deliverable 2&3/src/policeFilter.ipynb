```python
import pandas as pd
import numpy as np

%matplotlib inline

TARGET_FILES = [
    "data/Boston.csv",
    "data/Brockton.csv",
    "data/Cambridge.csv",
    "data/Lynn.csv",
    "data/Springfield.csv"
]
```

```python
city_infos = [{} for _ in TARGET_FILES]

for i in range(len(TARGET_FILES)):
    filename = TARGET_FILES[i]
    data = pd.read_csv(filename)
    data = data[data["Name"].notna()] # remove NaN name rows
    city_infos[i] = {
        "data": data,
        "filename": filename
    }
```

```python
# example of data
city_infos[0]["data"].head()
```

```python
def filter(df, col, pattern):
    mask = (df[col].astype(str).str.contains(pattern, case = False, regex = False)
            | df[col].isnull())
    return df[mask]

for i in range(len(city_infos)):
    df = city_infos[i]["data"]
    filename = city_infos[i]["filename"]
    city_infos[i]["data"] = filter(df, "Job_Title", "police")
    result_df = city_infos[i]["data"].reset_index(drop = True)
    result_df.to_csv(filename, index = False)

city_infos[4]["data"].head()
```
