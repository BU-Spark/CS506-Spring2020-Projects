```python
import pandas as pd
import numpy as np 
```

```python
boston_df = pd.read_csv("data/Boston/Boston.csv")
decertified_df = pd.read_csv("data/USA TODAY Police Decertifications List Search2020-01-13.csv")
decertified_df.rename(columns = {"Name:" : "Name"}, inplace = True)
```

```python
def nameMatch(caller, other, key):
    return caller.join(other.set_index(key), how = "inner", on = key)
```

```python
res = nameMatch(boston_df, decertified_df, key = "Name")
```
