```python
from PyPDF2 import *
import pandas as pd
import tabula
```

```python
worcester_2018_pdfs = tabula.read_pdf(r'.\data\Worcester\Worcester 2018.pdf', pages='all')
```

```python
worcester_2018_pdfs[4]
```

```python
for pdf in worcester_2018_pdfs:
    pdf.rename(columns={"Rank": "Rank Gross Pay", "Unnamed: 0": "Last Name", "Unnamed: 1": "First Name", "Unnamed: 2": "Job Title",
                       "Unnamed: 3": "Gross Pay", "Unnamed: 4": "Gross Pay No Pay Detail", "Unnamed: 5": "Regular Pay", 
                        "Unnamed: 6": "Pay Detail", "Unnamed: 7": "OT", "Unnamed: 8": "Nan", "Unnamed: 9": "WPS Aft Sch", "Unnamed: 10": "Court Att",
                       "Unnamed: 11": "Longevity Pay", "Unnamed: 12": "Principal", "Unnamed: 13": "Coach Pay", "Unnamed: 14": "HAZ MAT",
                       "Unnamed: 15": "Stipend", "Unnamed: 16": "Out of Grade", "Unnamed: 17": "Adj/Retro Other Pay", "Unnamed: 18": "NaN"}, inplace=True)
    pdf.drop('Nan', axis=1, inplace=True)
    if 'NaN' in pdf.columns:
        pdf.drop('NaN', axis=1, inplace=True)
```

```python
for pdf in worcester_2018_pdfs:
    pdf.drop([0, 1], axis=0, inplace=True)
```

```python
for names in  worcester_2018_pdfs[4]['Last Name']:
    name = names.split(' ')
    worcester_2018_pdfs[4]['Last Name'].loc[worcester_2018_pdfs[4]['Last Name']==names] = name[0]
    first_name = ''
    for i in range(1, len(name)):
        first_name = first_name + name[i] + ' '
#     first_name.strip(' ')
    worcester_2018_pdfs[4]['First Name'].loc[worcester_2018_pdfs[4]['Last Name']==name[0]] = first_name
```

```python
for names in  worcester_2018_pdfs[4]['Last Name']:
    name = names.split(' ')
    worcester_2018_pdfs[4]['Last Name'].loc[worcester_2018_pdfs[4]['Last Name']==names] = name[0]
    first_name = ''
    for i in range(1, len(name)):
        first_name = first_name + name[i] + ' '
#     first_name.strip(' ')
    worcester_2018_pdfs[4]['First Name'].loc[worcester_2018_pdfs[4]['Last Name']==name[0]] = first_name
```

```python
for pdf in worcester_2018_pdfs:
    pdf['Name'] = pdf['Last Name'] + ', ' + pdf['First Name']
```

```python
worcester_2018_pdf = pd.concat(worcester_2018_pdfs)
```

```python
worcester_2018_pdf = worcester_2018_pdf.set_index(['Rank Gross Pay'])
worcester_2018_pdf
```

```python
worcester_2014 = pd.read_csv(r'.\data\Worcester\Worcester2014.csv')
```

```python
worcester_2014 = worcester_2014[['Rank_Gross_Pay', 'Last_Name', 'First_Name', 'Job_Title', 'Gross_Pay', 'Regular_Pay', 'Pay_Detail', 'OT']]
worcester_2014.rename(columns={'Rank_Gross_Pay': 'Rank Gross Pay', 'Last_Name': 'Last Name', 'First_Name': 'First Name', 
                              'Job_Title': 'Job Title', 'Gross_Pay': 'Gross Pay', 'Regular_Pay': 'Regular Pay', 'Pay_Detail': 'Pay Detail',
                              'OT': 'OT'}, inplace=True)
worcester_2014 = worcester_2014.set_index(['Rank Gross Pay'])
worcester_2014
```

```python
for col in worcester_2014.columns.values:
    worcester_2014[col] = worcester_2014[col].str.strip()
worcester_2014
```

```python
dollars = ['Gross Pay', 'Regular Pay', 'Pay Detail', 'OT']
for dollar in dollars:
    worcester_2014[dollar] = worcester_2014[dollar].str.strip('$')

worcester_2014
```
