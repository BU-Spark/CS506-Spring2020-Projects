```python
import pandas as pd
import numpy as np
```

```python
'''
The list contains all the keywords of the titles that exist in the current police system of the US. 
Please refer to the details: https://online.wvstateu.edu/news/criminal-justice/u-s-police-ranking-system/
'''
policeList = ['police', 'officer', 'detective', 'corporal', 'sergeant', 'lieutenant', 'captain', 'inspector', 'commissioner']
```

```python
Boston_decertified = pd.read_csv(r'C:\Users\Memphis Meng\Desktop\Boston_decertified.csv')
```

```python
Boston_decertified = Boston_decertified.drop(['_type'], axis=1)
Boston_decertified
```

```python
mask = Boston_decertified.Job_Title.str.lower().apply(lambda x: any(item for item in policeList if item in x.split(' ')))
```

```python
Boston_decertified = Boston_decertified[mask]
Boston_decertified
```

```python
Boston_decertified.to_csv('Boston_decertified.csv')
```
